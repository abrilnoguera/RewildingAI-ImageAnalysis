```python
import boto3
import os
from tqdm import tqdm
import pandas as pd
```

```python
# Importo Informacion del AccessKey.
accessKey = pd.read_csv("ANoguera_accessKeys.csv")
```

```python
# Configura tus credenciales de AWS si aún no lo has hecho
boto3.setup_default_session(aws_access_key_id=accessKey["Access key ID"][0],
                           aws_secret_access_key=accessKey["Secret access key"][0],
                           region_name='us-east-2')
```

```python
# Define la ubicación de los archivos locales y el nombre del bucket S3
carpeta_local = "/Volumes/RewArg/Muestreos"
nombre_bucket_s3 = 's3-pfa'
prefijo_carpeta_s3 = "Muestreo ct sauce"
```

```python
# Crea un cliente de S3
s3 = boto3.client('s3')
```

```python
# Función para obtener una lista de todos los archivos locales en una carpeta y sus subcarpetas
def obtener_archivos_locales(carpeta):
    archivos = []
    for ruta, _, archivos_en_carpeta in os.walk(carpeta):
        for archivo in archivos_en_carpeta:
            archivos.append(os.path.join(ruta, archivo))
    return archivos
```

```python
# Lista los archivos locales
archivos_locales = obtener_archivos_locales(carpeta_local)
```

```python
archivos_locales
```

```python
# archivos_para_cargar
archivoslocales = []
for archivo in archivos_locales:
    partes_ruta = archivo.split(os.path.sep)
    archivoslocales.append(os.path.join(*partes_ruta[4:]))

del(archivos_locales)
```

```python
archivoslocales
```

```python
len(archivoslocales)
```

```python
paginator = s3.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=nombre_bucket_s3)
objetos_s3 = []

for page in pages:
    for obj in page['Contents']:
        objetos_s3.append(obj['Key'])
```

```python
len(objetos_s3)
```

```python
objetos_s3
```

```python
# Filtra los archivos que no existen aún en S3
archivos_para_cargar = [archivo for archivo in archivoslocales if archivo not in objetos_s3]
```

```python
archivos_para_cargar
# 'Muestreo ct sauce/CT S 2022/S16/1/IMG_0001.JPG'
```

```python
len(archivos_para_cargar)
```

```python
# Si hay archivos para cargar
if archivos_para_cargar:
    print(f"Cargando {len(archivos_para_cargar)} archivos a S3...")
    for archivo in tqdm(archivos_para_cargar):
        archivo_local = os.path.join(carpeta_local, archivo)
        archivo_s3 = archivo_local.replace(carpeta_local, '').lstrip('/')
        archivo_s3 = archivo_s3.replace("  "," ")
        s3.upload_file(archivo_local, nombre_bucket_s3, archivo_s3)
    print("Carga completa.")
else:
    print("No hay archivos nuevos para cargar a S3.")
```

Elimino los espacios adelante de los nombres del archivo:

```python
# Recorrer cada objeto y renombrarlos eliminando espacios al principio
for objeto in objetos_s3:
    clave_original = objeto
    partes = clave_original.split('/')  # Dividir la clave en partes
    partes_sin_espacios = [parte.strip(' ') for parte in partes]  # Eliminar espacios en blanco después de cada "/"
    clave_sin_espacios = '/'.join(partes_sin_espacios)  # Volver a unir las partes en una nueva clave
    if clave_sin_espacios != clave_original:
        # Solo renombrar si la clave se modificó
        s3.copy_object(Bucket=nombre_bucket_s3, CopySource={'Bucket': nombre_bucket_s3, 'Key': clave_original},
                       Key=clave_sin_espacios)
        s3.delete_object(Bucket=nombre_bucket_s3, Key=clave_original)

print("Archivos renombrados exitosamente.")
```

Elimino espacios dobles:

```python
# Recorrer cada objeto y renombrar eliminando espacios dobles al principio de la clave
for objeto in objetos_s3:
    clave_original = objeto
    clave_sin_espacios = clave_original.replace('  ', ' ')  # Elimina espacios dobles al principio
    if clave_sin_espacios != clave_original:
        print(clave_sin_espacios)
        # Solo renombrar si hay espacios dobles al principio
        s3.copy_object(Bucket=nombre_bucket_s3, CopySource={'Bucket': nombre_bucket_s3, 'Key': clave_original},
                       Key=clave_sin_espacios)
        s3.delete_object(Bucket=nombre_bucket_s3, Key=clave_original)

print("Archivos renombrados exitosamente.")
```
